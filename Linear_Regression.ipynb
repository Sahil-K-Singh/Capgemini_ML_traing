```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('House_price_prediction.csv')
```

```python
df
```

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OneHotEncoder
```

```python
label=LabelEncoder()
df['Type_encoded']=label.fit_transform(df['Type'])
```

```python
df
```

```python
df['method_encoded']=label.fit_transform(df['Method'])
df
```

```python
df['region_encoded']=label.fit_transform(df['Regionname'])
df
```

```python
#pd.get_dummies(df,dtypes=int)
```

```python
df.drop(columns=['Type','Method','Regionname'],axis=1,inplace=True)
```

```python
df
```

```python
Onehot=OneHotEncoder(sparse_output=False)
suburb_data=Onehot.fit_transform(df[['Suburb']])
suburb_data
s_name = Onehot.get_feature_names_out(['Suburb'])  
s_encoded= pd.DataFrame(suburb_data, columns=s_name)
s_encoded.shape
```

```python
council=Onehot.fit_transform(df[['CouncilArea']])
c_name = Onehot.get_feature_names_out(['CouncilArea'])  
c_encoded= pd.DataFrame(council, columns=c_name)
c_encoded
```

```python
seller=Onehot.fit_transform(df[['SellerG']])
se_name = Onehot.get_feature_names_out(['SellerG'])  
se_encoded= pd.DataFrame(seller, columns=se_name)
se_encoded
```

```python
df=pd.concat([df,se_encoded,c_encoded,s_encoded],axis=1)
df
```

```python
df.drop(columns=['Suburb','SellerG','CouncilArea'],axis=1,inplace=True)
```

```python
df
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score,mean_squared_error,mean_absolute_error
```

```python
x=df.drop('Price',axis=1)
```

```python
y=df['Price']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.3,random_state=42)
```

```python
model=LinearRegression()
```

```python
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
y_pred
```

```python
model.score(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

```python
r2=r2_score(y_test,y_pred)
```

```python
r2
```

```python
mse=mean_squared_error(y_test,y_pred)
mse
```

```python
mae=mean_absolute_error(y_test,y_pred)
```
