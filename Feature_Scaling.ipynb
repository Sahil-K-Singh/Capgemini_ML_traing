```python
import pandas as pd
data={'sname':['s1','s2','s3','s4'],
     'height':[162,171,155,165],
     'weight':[63,60,50,62],
     'speed':[5,3,7,2]}
df=pd.DataFrame(data)
df
```

```python
from sklearn.preprocessing import Normalizer
```

```python
norm=Normalizer()
df_num=df[['height','weight','speed']]
data=norm.fit_transform(df_num)
df=pd.DataFrame(data,columns=df_num.columns)
df
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
ss=StandardScaler()
df_num=df[['height','weight','speed']]
data=ss.fit_transform(df_num)
df=pd.DataFrame(data,columns=df_num.columns)
df
```
